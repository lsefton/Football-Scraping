```python
from time import time
import pandas as pd
from unidecode import unidecode
import re


def merge_df(df_fpl, df_understat):
    #find new indices in understat data for players that match fpl data
    new_indices = []
    all_indices = range(len(df_fpl['name']))
    for u_index in range(len(df_understat['name'])):
        for f_index in range(len(df_fpl['name'])):
            if re.search(r'\b' + df_fpl['name'][f_index].lower() + r'\b', df_understat['name'][u_index].lower()) and df_fpl['team'][f_index] == df_understat['team'][u_index]:
                new_indices.append(f_index)
                break
    #create new id columns
    df_understat['id'] = new_indices
    df_fpl['id'] = range(len(df_fpl['name']))
    #merge on new id
    df_merged = pd.merge(df_fpl, df_understat, on='id', suffixes=('', '_y'))
    #remove duplicate columns
    df_merged.drop(list(df_merged.filter(regex = '_y')), axis = 1, inplace = True)
    df_merged.drop(list(df_merged.filter(regex = 'id')), axis = 1, inplace = True)
    
    return df_merged


### start of code, general initialisation ###
start_time = time()
year = 2018
data_path = r"C:\Users\Luke Sefton\Documents\Web scraping\FPL\data\%i-%i" % (year, year+1)

df_understat = pd.read_csv(r'%s\understat_EPL_%i.txt' % (data_path, year))
df_fpl = pd.read_csv(r'%s\fpl_players_%i.txt' % (data_path, year))

print('Time: %.2fs' % (time()-start_time))

print(len(df_fpl), len(df_understat))

#df_fpl
```

```python
df_merged = merge_df(df_fpl, df_understat)

field = 'npxG'
team = 'TOT'
position = 'D'
min_minutes = 45
df_sorted = df_merged.sort_values(by=[field], ascending=False)
df_filtered = df_sorted.loc[(df_merged['minutes'] >= min_minutes)
                            #& (df_merged['team'] == team)
                            #& (df_merged['position'] == position)
                           ]
df_filtered.style.format({'price': "£{:.1f}m"})

        
        
```
